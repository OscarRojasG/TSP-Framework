```python
from src.training import load_data

dataset = load_data(file_path="TSP50_data.pkl")
```

```python
# FASE 1: Fijamos parÃ¡metros de arquitectura
# LR, batch size y dropout fijos
from src.models.transformer_v5 import TSPTransformer
from src.optuna import *

lr = OptunaLR(1e-4, 1e-4)
batch_size = OptunaBatchSize([32])
hyperparams = OptunaHyperparams(lr, batch_size)

hyperparams.register(OptunaCategorical("embed_dim", [32, 64, 128]))
hyperparams.register(OptunaCategorical("num_heads", [2, 4, 8]))
hyperparams.register(OptunaCategorical("num_encoder_layers", [1, 2, 4]))
hyperparams.register(OptunaCategorical("num_glimpses", [1, 2, 4]))
hyperparams.register(OptunaCategorical("dropout_rate", [0.1]))

study = optimize_hyperparams(
    TSPTransformer,
    hyperparams,
    n_trials=20,
    dataset=dataset,
    epochs=50,
    train_size=4000,
    test_size=1000,
)
```
